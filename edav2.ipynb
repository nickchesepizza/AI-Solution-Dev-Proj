#TEMPERATURE

```python
print("Descriptive statistics for 'Temperature' column:")
print(df['Temperature'].describe())

print("\nMissing values in 'Temperature' column:")
print(df['Temperature'].isnull().sum())
```

Things taken note
1. Standard deviation is crazy, 74 degress celcius of houses is unreasonable.
2. max and min values is too unreasonable for housing
3. looking at the percentiles, majority of the data is between 18.9 to 23.1 degrees

Find acceptable range (IQR METHOD)
1. IQR is Q3 - Q1 (23.17 - 18.94 = 4.23) IQR = 4.23
2. Upper bound is 23.17 + (1.5 * 4.23) = 29.52
3. Lower bound is 18.94 - (1.5 * 4.23) = 12.59
4. acceptable range is 12.59 to 29.52

```python
print(f"Original rows: {len(df)}")

lower_bound = 12.59
upper_bound = 29.52

df = df[(df['Temperature'] >= lower_bound) & (df['Temperature'] <= upper_bound)].copy()

print(f"Rows after filtering Temperature: {len(df)}")
```

To clean the HVAC Operation Mode data, we first converted all the text to uppercase so that the categories matched perfectly. Because machine learning models need numbers instead of words, we used One-Hot Encoding to turn each HVAC mode into its own column of 0s and 1s. Finally, we used the drop_first=True setting to remove one of the duplicate categories. This prevents a mathematical error called 'multicollinearity' (the dummy variable trap), which would otherwise confuse our Logistic Regression model."

#Ambient Light Level

```python
print("Descriptive statistics for 'Ambient Light Level' column:")
print(df['Ambient Light Level'].describe())

print("\nMissing values in 'Ambient Light Level' column:")
print(df['Ambient Light Level'].isnull().sum())

print("\nValue counts for 'Ambient Light Level' column:")
print(df['Ambient Light Level'].value_counts())
```

```python
print(f"Original rows before dropping NaNs in 'Ambient Light Level': {len(df)}")
df.dropna(subset=['Ambient Light Level'], inplace=True)
print(f"Rows after dropping NaNs in 'Ambient Light Level': {len(df)}")
display(df.head())
```

```python
light_mapping = {
    'very_dim': 0,
    'dim': 1,
    'moderate': 2,
    'bright': 3,
    'very_bright': 4
}

df['Ambient Light Level'] = df['Ambient Light Level'].map(light_mapping)
```

The Ambient Light Level feature presented as ordinal categorical data with 594 missing values. We just dropped them. Because the categories possess a definitive inherent ranking, Ordinal Encoding was utilized rather than One-Hot Encoding. The categories were mapped to numerical integers (0 to 4), preserving the linear relationship of light intensity for the baseline Logistic Regression model.

#Activity Level

```python
print("Descriptive statistics for 'Activity Level' column:")
print(df['Activity Level'].describe())

print("\nMissing values in 'Activity Level' column:")
print(df['Activity Level'].isnull().sum())

print("\nValue counts for 'Activity Level' column:")
print(df['Activity Level'].value_counts())
```

```python
target_mapping = {
    'Low Activity': 0,
    'Low_Activity': 0,
    'LowActivity': 0,
    'Moderate Activity': 1,
    'ModerateActivity': 1,
    'High Activity': 2
}


df['Activity Level'] = df['Activity Level'].map(target_mapping)
```

Analysis of the target variable, Activity Level, revealed data entry inconsistencies resulting in duplicated categorical classes (e.g., 'Low Activity', 'Low_Activity'). A mapping dictionary was utilized to consolidate these typographic errors and simultaneously apply Ordinal Encoding, transforming the target into a numerical scale (0 for Low, 1 for Moderate, 2 for High). Post-consolidation, a class imbalance was identified, with 'High Activity' representing only 11% of the dataset. This highlights the necessity of evaluating subsequent predictive models using Recall and F1-scores rather than baseline accuracy, and integrating class-weight balancing during model training
